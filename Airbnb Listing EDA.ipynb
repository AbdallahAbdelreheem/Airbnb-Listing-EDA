# Data Cleaning
Handle missing values, remove duplicates, and correct data types if necessary.

```python

# Remove duplicates
df.drop_duplicates(inplace=True)

# Correct data types if necessary
df['price'] = df['price'].astype(str).str.replace('$', '').str.replace(',', '').astype(float)
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)

# Fill missing values before converting to integers
df['accommodates'] = df['accommodates'].fillna(0).astype(int)
df['bathrooms'] = df['bathrooms'].fillna(0).astype(float)
df['beds'] = df['beds'].fillna(0).astype(int)
df['minimum_nights'] = df['minimum_nights'].fillna(0).astype(int)
df['maximum_nights'] = df['maximum_nights'].fillna(0).astype(int)
df['availability_30'] = df['availability_30'].fillna(0).astype(int)
df['availability_60'] = df['availability_60'].fillna(0).astype(int)
df['availability_90'] = df['availability_90'].fillna(0).astype(int)
df['availability_365'] = df['availability_365'].fillna(0).astype(int)
df['number_of_reviews'] = df['number_of_reviews'].fillna(0).astype(int)
df['number_of_reviews_ltm'] = df['number_of_reviews_ltm'].fillna(0).astype(int)
df['number_of_reviews_l30d'] = df['number_of_reviews_l30d'].fillna(0).astype(int)

df['review_scores_rating'] = df['review_scores_rating'].astype(float)
df['review_scores_accuracy'] = df['review_scores_accuracy'].astype(float)
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].astype(float)
df['review_scores_checkin'] = df['review_scores_checkin'].astype(float)
df['review_scores_communication'] = df['review_scores_communication'].astype(float)
df['review_scores_location'] = df['review_scores_location'].astype(float)
df['review_scores_value'] = df['review_scores_value'].astype(float)
df['reviews_per_month'] = df['reviews_per_month'].astype(float)

# Display the cleaned dataframe
df.head()
```

# Exploratory Data Analysis
Perform exploratory data analysis to understand the distribution and relationships in the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

```python
# Plot the distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of review scores rating
plt.figure(figsize=(10, 6))
sns.histplot(df['review_scores_rating'], bins=50, kde=True)
plt.title('Distribution of Review Scores Rating')
plt.xlabel('Review Scores Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the relationship between price and accommodates
plt.figure(figsize=(10, 6))
sns.scatterplot(x='accommodates', y='price', data=df)
plt.title('Price vs Accommodates')
plt.xlabel('Accommodates')
plt.ylabel('Price')
plt.show()
```

```python

# Plot the relationship between price and number of reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=df)
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```
